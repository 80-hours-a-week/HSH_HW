# 반복문 - 미니퀘스트 & 과제

## 1번 퀘스트

```python
RepeatCount = input("숫자를 입력:")
if RepeatCount < 2:
  print("2 이상의 숫자를 입력하세요")
else:
  for i in range(RepeatCount):
    ResultCount = RepeatCount*i
    print(f"{RepeatCount} x {i} = {ResultCount}")
```

## 2번 퀘스트

```python
n = 5  # 삼각형의 높이 (밑변이 9일 때 높이는 5)

for i in range(1, n + 1):
    # 공백 출력: 높이 - 현재 층수
    # 별 출력: 2 * 현재 층수 - 1
    print(' ' * (n - i) + '*' * (2 * i - 1))
# by Gemini
```

# Week 1 과제 - 1차 수정

```python
  # 몬스터의 초기 체력
  monster_hp = 1000

  print("몬스터를 토벌합니다!")
  print(f"몬스터의 초기 체력: {monster_hp}")

  # 몬스터 체력이 0보다 클 동안 반복
  while monster_hp > 0:
      print(f"\n현재 몬스터 체력: {monster_hp}")
      print("사용할 스킬을 선택하세요:")
      print("1. 화염 (데미지: 300)")
      print("2. 물   (데미지: 200)")
      print("3. 풀   (데미지: 100)")

      choice = input("스킬 번호 입력 (1/2/3): ")

      skills = {
          '1': ("화염", 300),
          '2': ("물", 200),
          '3': ("풀", 100)
      }

      if choice not in skills:
          print("잘못된 입력입니다.")
          continue

      skill_name, damage = skills[choice]

      # 공격 프로세스 + 몬스터 체력 관리 함수
      def monsterhp(monster_hp):
        monster_hp -= damage
        if monster_hp < 0:
          monster_hp = 0
        return monster_hp

      monster_hp = monsterhp(monster_hp)

      print(f"몬스터에게 {damage}의 데미지를 입혔습니다 (남은 체력: {monster_hp})")

      # 몬스터 생존 여부 확인 후 반격 출력
      if monster_hp > 0:
          print(" 몬스터가 반격했다. 그러나 아무 일도 없었다.")
      else:
          print("몬스터 처치 완료!")
          break
```
